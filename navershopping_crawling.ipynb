```python
from selenium import webdriver
from bs4 import BeautifulSoup
from urllib.request import urlretrieve
import urllib.request
import os
import requests
import time
from random import uniform
import pandas as pd
```

```python
##### chromedriver로 크롬 사이트 접속하기
driver = webdriver.Chrome('c:/chromedriver')
driver.implicitly_wait(3)
# 네이버 쇼핑윈도 - 원피스 카테고리로 들어가기
driver.get('https://search.shopping.naver.com/search/category.nhn?pagingIndex=1&pagingSize=40&productSet=window&viewType=list&sort=rel&cat_id=50000807&frm=NVSHCAT')
```

```python
link_list = []

try:
    for page in range(1):
        # 페이지에 있는 상품 url 저장
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        links = soup.select('div.img_area > a')
        f = uniform(2.0, 2.5)
        
        for link in links:
            link_list.append(link.get('href'))
        
        # 다음페이지 클릭
        driver.find_element_by_class_name('next').click()
        time.sleep(0.5)

except:
    print("더 이상 페이지가 없습니다.")
```

```python
# csv 파일로 저장
results = pd.DataFrame(link_list, columns=['link_addr'])
results.to_csv('./link_addr.csv', header=False) #, mode='a', header
```
