```python
#색 구별 하지않은 파일
path = "./one/data/" #파일경로
file_list = os.listdir(path)
file_list_csv = [file for file in file_list if file.endswith(".csv")]#csv 만
#print ("file_list: {}".format(file_list_csv))
#file_list_csv
```

```python
for file in file_list_csv[1:]: #첫 파일이 전체 리스트이므로 1 부터
    f = open('./one/data/'+file, 'r', encoding='utf-8')
    rdr = csv.reader(f)
    next(rdr)
    for line in rdr:
        line[9]=re.sub('[/]','',line[9]) #폴더생성에 방해되는 슬레시 제거
        if "기타" in [line[4],line[5],line[6],line[9]]: #속성에 기타가 포함되면 작업 하지 않음
            pass
        else:
            try:
                if not(os.path.isdir('./one/'+line[4]+'_'+line[5]+'_'+line[6]+'_'+line[9])): #속성명으로 만들어진 폴더가 있는지 확인
                    os.makedirs(os.path.join('./one/'+line[4]+'_'+line[5]+'_'+line[6]+'_'+line[9]))#속성명으로 폴더 생성
                else:
                    print('이미존재하는디렉터리')        
            except OSError as e:
                if e.errno != errno.EEXIST:
                    print("Failed to create directory!!!!!")
                    raise
            shutil.copy('./one/data/'+line[11],'./one/'+line[4]+'_'+line[5]+'_'+line[6]+'_'+line[9]+'/')#원폰 그림파일을 분류도니 폴더로
            #print('./one/data/'+line[11],'./one/'+line[4]+'_'+line[5]+'_'+line[6]+'_'+line[9]+'/')#메시지 표시
f.close() 
    
```
